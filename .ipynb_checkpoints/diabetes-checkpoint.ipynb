### UNIVARIATE ANALYSIS.

#### Numerical Features.

```python
# Get a list of numerical variables without the ID column.

Numerical_variables = df.select_dtypes(include=['number']).columns.tolist()[1:]
Numerical_variables
```

```python
# Distribution of all numerical variables using histogrm.
Numerical_variables

#Set the figure size and plot the figure.
df[Numerical_variables].hist(figsize=(16,12), bins=30, edgecolor='black')
plt.tight_layout()
plt.suptitle('Distribution of Numerical Variables', fontsize=16, y=1.02)
plt.show()
```

```python
#Plot KDE for each numerical variable
#Numerical_variables

#Set up the plot grid.
plt.figure(figsize=(18,14))

#Plot KDE for each numerical variable

for i, var in enumerate(Numerical_variables, 1):
    plt.subplot((len(Numerical_variables)+2)//3, 3, i)
    sns.kdeplot(data=df, x=var, fill=True, color='skyblue')
    plt.title(f'Distribution (KDE) of {var}')
    plt.xlabel(var)
    plt.ylabel('Density')

plt.tight_layout()
plt.suptitle('KDE Plots of NUmerical Variables', fontsize=18, y=1.02)
plt.show()
```

#### Visualize Outliers

```python
# Distribution of all numerical variables using boxplot to visualize outliers.
Numerical_variables

#Plot Boxplot for each numerical variable

#Set up the plot grid.
plt.figure(figsize=(18,14))

for i, var in enumerate(Numerical_variables, 1):
    plt.subplot((len(Numerical_variables)+2)//3, 3, i)
    sns.boxplot(data=df, y=var, color='skyblue')
    plt.title(f'Distribution (Boxplot) of {var}')

plt.tight_layout()
plt.suptitle('Box Plots of Numerical Variables', fontsize=18, y=1.02)
plt.show()
```

#### Compute Outliers.

```python
# Compute the outliers with z-score for normal distribution ONLY.
# The variables in this dataset have skewed distribution.
# Alternatively, use IQR method compute the outliers.

# Dictionary to store outliers for each varible.
outliers_dict = {}

for var in Numerical_variables:
    Q1 = df[var].quantile(0.25)
    Q3 = df[var].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 - 1.5 * IQR

    # Find Outliers.
    outliers = df[(df[var] < lower_bound) | (df[var] > upper_bound)]

    # Store Results
    outliers_dict[var] = {
        'count': outliers.shape[0],
        'indices': outliers.index.tolist(),
        'values': outliers[var].tolist()
    }

# Display Outlier Summary
for feature, stats in outliers_dict.items():
    print(f'{feature} -> Outliers: {stats['count']}, {stats['values']}')
```

```python

import os

#Create a directory to save the plots
output_dir = 'kde plots'
os.makedirs(output_dir, exist_ok=True)

#Generate and save each kde plot

for var in Numerical_variables:
    plt.figure(figsize=(8,6))
    sns.kdeplot(data=df, x=var, fill=True, color='skyblue')
    plt.title(f'Distribution (KDE) of {var}')
    plt.xlabel(var)
    plt.ylabel('Density')
    plt.tight_layout()

    #save the figure.
    file_path = os.path.join(output_dir, f'{var}_kde.png')
    plt.savefig(file_path)
    plt.close() #Close the figure to save memory.

print('completed')
```

```python
# group ages into caegories.

# below 31, 31-40, 41-50, 51-60, 61-70, 71-80, above 80.

def getAgeGroup(age):
    if age < 31: return 'below 31'
    elif age <41: return '31-40'
    elif age <51: return '41-50'
    elif age <61: return '51-60'
    elif age <71: return '61-70'
    elif age <81: return '71-80'
    else: return 'above 80'
```

```python
df['Age_Group'] = df['Age'].apply(lambda x: getAgeGroup(x))
```

```python
df['Age_Group']
```

```python
# Set the orders for the Age_Group
# - to reduce memory usage
# - for meaningful data analysis and visualization.

# Import the class.
from pandas.api.types import CategoricalDtype

Age_Group_levels = ['below 31', '31-40', '41-50', '51-60', '61-70', '71-80', 'above 80']
Age_Group_Levels_order = CategoricalDtype(categories=Age_Group_levels, ordered=True)

# Set the orders for the Age_Group

df["Age_Group"] = df["Age_Group"].astype(Age_Group_Levels_order)
```
